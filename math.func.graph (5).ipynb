```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
## sin(), cos()
x = np.arange(0, 2*math.pi, 0.1);

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
## logistic regression function

x = np.arange(-5, 5, 0.3);
y = 1/(1+np.exp(-x))

plt.plot(x, y)
```

```python
## standard normal distribution 

# python math operators:
# https://www.w3schools.com/python/python_operators.asp

# the normal distribution probability density function (PDF)
# https://www.itl.nist.gov/div898/handbook/eda/section3/eda3661.htm


def snorm(x):
    return math.e**(-x**2/2)/(math.sqrt(2*math.pi))

x = np.arange(-5, 5, 0.1);
plt.plot(x, snorm(x))

#y = (1/(2*math.pi))**(math.exp(-1*x**2)/2)
```

```python
## standard normal distribution - cumulative distribution function (CDF)

# the normal distribution cumulative distribution function (CDF)
# https://www.itl.nist.gov/div898/handbook/eda/section3/eda3661.htm
# ... CDF is the integral (-unlimimted, x) of the PDF 


# the erf - error function, and more special functions
# https://docs.python.org/3.2/library/math.html

'''
def phi(x):
    # standard normal CDF
    #return math.erf(x / math.sqrt(2.0))
    return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0
'''

# x is a range
def phi(x):
    # standard normal CDF
    #return math.erf(x / math.sqrt(2.0))
    x1= x / math.sqrt(2.0)
    #print(x1)
    
    #x2= map(lambda e: math.erf(e), x1)
    x2 = [(1.0+math.erf(x))/2.0 for x in x1]
    
    #print(x2)
        
    
    return x2

x = np.arange(-5, 5, 0.1)


#print(x)
plt.plot(x, phi(x))



# read more:
# a list of distribution functions
# https://support.minitab.com/en-us/minitab-express/1/help-and-how-to/basic-statistics/probability-distributions/how-to/cumulative-distribution-function-cdf/methods-and-formulas/methods-and-formulas/
```

```python
#two tail CI 

def cdf(x):
    # standard normal CDF
    #return math.erf(x / math.sqrt(2.0))
    return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0

print(cdf(1.96))    
```
