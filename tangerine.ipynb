# TimeSeries and trends plot

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
%matplotlib inline
# sns.set()
```

```python
df = pd.read_csv('tangerine.txt', skiprows=0, sep='\t')
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df.head()
```

```python
df.daily_nav_history = pd.to_datetime(df.daily_nav_history, format='%B %d, %Y')
```

```python
df.head()
```

```python
df.set_index('daily_nav_history', inplace=True)
df.head()
```

```python
df['balanced_income'] = df['balanced_income'].str.replace("$","")
df['balanced'] = df['balanced'].str.replace("$","")
df['balanced_growth'] = df['balanced_growth'].str.replace("$","")
df['dividend'] = df['dividend'].str.replace("$","")
df['equity_growth'] = df['equity_growth'].str.replace("$","")



df['balanced_income'] = pd.to_numeric(df['balanced_income'])
df['balanced'] = pd.to_numeric(df['balanced'])
df['balanced_growth'] = pd.to_numeric(df['balanced_growth'])
df['dividend'] = pd.to_numeric(df['dividend'])
df['equity_growth'] = pd.to_numeric(df['equity_growth'])


df.head()
df.info()
```

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```
